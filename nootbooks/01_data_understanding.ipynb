```python
from sentence_transformers import SentenceTransformer

bert_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
product_texts = product_reviews['combined_reviews'].tolist()

bert_embeddings = bert_model.encode(product_texts, show_progress_bar=True)

bert_embeddings.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
bert_similarity = cosine_similarity(bert_embeddings)
```

```python
def get_bert_recommendations(product_name, top_n=5):
    product_index = product_reviews[
        product_reviews['name'].str.lower() == product_name.lower()
    ].index[0]

    similarity_scores = list(enumerate(bert_similarity[product_index]))

    sorted_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    top_products = sorted_scores[1:top_n+1]

    recommended_indices = [i[0] for i in top_products]

    return product_reviews[['name']].iloc[recommended_indices]
```

```python
def get_tfidf_recommendations(product_name, top_n=5):
    product_index = product_reviews[
        product_reviews['name'].str.lower() == product_name.lower()
    ].index[0]

    similarity_scores = list(enumerate(cosine_sim[product_index]))

    sorted_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    top_products = sorted_scores[1:top_n+1]

    recommended_indices = [i[0] for i in top_products]

    return product_reviews[['name']].iloc[recommended_indices]
```

```python
hybrid_similarity = (cosine_sim + bert_similarity) / 2
```

```python
def recommend_products_hybrid(product_name, top_n=5):
    # find products that contain the text
    matches = df[
        df["name"].str.lower().str.contains(product_name.lower(), na=False)
    ]

    # if no product is found
    if len(matches) == 0:
        return "Product not found. Try another product name."

    # use the first matching product
    product_index = matches.index[0]

    # get hybrid similarity scores
    similarity_scores = list(enumerate(hybrid_similarity[product_index]))

    # sort scores
    similarity_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    # remove same product
    similarity_scores = similarity_scores[1:top_n + 1]

    # get indexes
    recommended_indexes = [i[0] for i in similarity_scores]

    return df.iloc[recommended_indexes][["name"]]
```

```python
def recommend_products_hybrid(product_name, top_n=5):
    matches = df[
        df["name"].str.lower().str.contains(product_name.lower(), na=False)
    ]

    if len(matches) == 0:
        return "Product not found. Try another product name."

    product_position = matches.index[0]

    similarity_scores = list(enumerate(hybrid_similarity[product_position]))

    similarity_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    similarity_scores = similarity_scores[1:top_n + 1]

    recommended_positions = [i[0] for i in similarity_scores]

    return df.iloc[recommended_positions][["name"]]
```
